```python
for year in years:
    data = requests.get(standings_url)
    soup = BeautifulSoup(data.text)
    standings_table1 = soup.select('table.stats_table')[0]
    standings_table2 = soup.select('table.stats_table')[1]

    links1 = standings_table1.find_all('a')
    links2 = standings_table2.find_all('a')

    both_set_of_links = [l.get('href') for l in links1] + [l.get('href') for l in links2]
    team_links = [l for l in both_set_of_links if '/teams/' in l]
    
    team_urls = [f"https://www.pro-football-reference.com{l}" for l in team_links]

    previous_season = soup.select("a.prev")[0].get("href")
    standings_url = f"https://www.pro-football-reference.com{previous_season}"

    for team_url in team_urls:
        team_name = team_url.split("/")[-2]
    
        data = requests.get(team_url)
        games = pd.read_html(data.text, match="Schedule & Game Results")
    
        # Check if the games DataFrame has a MultiIndex in its columns
        if isinstance(games.columns, pd.MultiIndex):
            games.columns = games.columns.droplevel(0)  # Only drop the first level if it's a MultiIndex
    
        soup = BeautifulSoup(data.text)
        links = [l.get("href") for l in soup.find_all('a')]
        links = [l for l in links if l and '/gamelog/' in l]
        data = requests.get(f"https://www.pro-football-reference.com{links[0]}")
        opponent_logs = pd.read_html(data.text, match="Opponent Game Logs")[0]
    
        # Again, check if opponent_logs DataFrame has a MultiIndex
        if isinstance(opponent_logs.columns, pd.MultiIndex):
            opponent_logs.columns = opponent_logs.columns.droplevel(0)
    
        games['Week'] = games['Week'].astype(str)
        opponent_logs['Week'] = opponent_logs['Week'].astype(str)
    
        try:
            team_data = games.merge(opponent_logs[["Week", "Cmp%", "ToP"]], on="Week")
        except ValueError:
            continue
    
        team_data["Season"] = year
        team_data["Team"] = team_name
        all_games.append(team_data)
        time.sleep(1)
```

```python
game_df = pd.concat(all_games)
```

```python
game_df.to_csv("games.csv")
```

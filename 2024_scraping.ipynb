```python
for team_url in team_urls:
    team_name = team_url.split("/")[-2]
    
    # Fetch team schedule
    team_data_response = requests.get(team_url)
    team_soup = BeautifulSoup(team_data_response.text, 'html.parser')
    
    # Extract schedule and game results
    future_games_list = pd.read_html(team_data_response.text, match="Schedule & Game Results")
    future_games = future_games_list[0]
    
    # Check if the games DataFrame has a MultiIndex in its columns
    if isinstance(future_games.columns, pd.MultiIndex):
        future_games.columns = future_games.columns.droplevel(0)
    
    # Parse team URLs for opponent logs
    links = [l.get("href") for l in team_soup.find_all('a')]
    opponent_links = [l for l in links if l and '/gamelog/' in l]
    
    # Process each opponent log link
    for opponent_link in opponent_links:
        opponent_log_url = f"https://www.pro-football-reference.com{opponent_link}"
        opponent_log_response = requests.get(opponent_log_url)
        opponent_logs_list = pd.read_html(opponent_log_response.text, match="Opponent Game Logs")
        
        # Check if opponent_logs DataFrame has a MultiIndex
        if isinstance(opponent_logs_list[0].columns, pd.MultiIndex):
            opponent_logs_list[0].columns = opponent_logs_list[0].columns.droplevel(0)
        
        # Ensure 'Week' columns are strings
        future_games['Week'] = future_games['Week'].astype(str)
        opponent_logs_list[0]['Week'] = opponent_logs_list[0]['Week'].astype(str)
        
        # Merge data on 'Week' column
        try:
            team_data = future_games.merge(opponent_logs_list[0][["Week", "Cmp%", "ToP"]], on="Week")
            team_data["Season"] = year
            team_data["Team"] = team_name
            all_2024_games.append(team_data)
        except ValueError:
            print(f"Error merging for {team_name}")
            continue  # Skip merging if there's an error

    time.sleep(10)  # Respectful pause between requests
```
